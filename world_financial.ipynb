```python
import statsmodels.api as sm

# Create a DataFrame from your data
data2 = list(actual)
data1 = [83.03510176, 85.96970412, 90.27686886, 101.3151649, 115.4876083, 119.0824759, 89.75758261, 96.04632976, 99.51683604,
        107.182143, 111.9683181, 118.1605353, 122.6124534, 143.456125, 163.2316156, 157.929385, 161.1372361,
        186.4190959, 206.0737493, 224.5754377, 267.3905787, 271.4261559, 275.2660979, 292.6446583, 278.0954074,
        297.9992568, 312.0598439, 342.0719237, 355.738418, 347.4620369, 368.7497594, 303.850438, 317.5587347,
        301.5007864, 346.2273931, 373.6282338, 399.5773066, 414.8986835, 412.5093558, 440.9614546, 442.0347789,
        449.9111249, 468.8444283, 543.8437989, 624.1050944, 710.5093449, 802.013742, 1022.732476, 993.5034125,
        1096.636136, 1350.634457, 1449.603301, 1434.017978, 1438.056995, 1559.863772, 1590.174322, 1714.279541,
        1957.969814, 1974.37773, 2050.16379, 1913.219721, 2238.127139, 2388.621198]


# Convert the data to a Pandas Series with a datetime index (assuming yearly data)
years =range(1960, 2023)
gdp_series = pd.Series(data1, index=pd.to_datetime(years, format='%Y'))

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(gdp_series)
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita Time Series')
plt.grid(True)
plt.show()

# Decompose the time series to identify trends and seasonality
decomposition = sm.tsa.seasonal_decompose(gdp_series, model='additive')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(gdp_series, label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()

# Train a time series forecasting model (e.g., SARIMA)
# Here, you can use AutoARIMA to automatically select the best model parameters
from pmdarima.arima import auto_arima
model = auto_arima(gdp_series, seasonal=True, m=12, stepwise=True, trace=True, suppress_warnings=True)
print(model.summary())

# Fit the SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX
order = model.order
seasonal_order = model.seasonal_order
sarima_model = SARIMAX(gdp_series, order=order, seasonal_order=seasonal_order)
sarima_results = sarima_model.fit()

# Forecast GDP per capita for the next five years (10 years)
forecast_steps = 10
forecast = sarima_results.get_forecast(steps=forecast_steps)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(gdp_series, label='Observed', color='blue')
plt.plot(forecast_mean.index, forecast_mean.values, color='red', label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita Forecast for the Next 10 Years')
plt.legend()
plt.grid(True)
plt.show()

# Print the forecasted values for the next 5 years
forecasted_values = forecast_mean[-10:]
print("Forecasted GDP per Capita for the Next 10 Years:")
print(forecasted_values)
```
